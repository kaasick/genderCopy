```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("average_by_country.csv")

# Select a subset of indicators for the plot
indicators_to_plot = [
    'Cost of business start-up procedures, female (% of GNI per capita)',
    'Cost of business start-up procedures, male (% of GNI per capita)'
]

# Sum values across all countries for each indicator
sum_female_education = df[indicators_to_plot[0]].sum()
sum_male_education = df[indicators_to_plot[1]].sum()

# Create a DataFrame for the sums
df_sums = pd.DataFrame({
    'Indicator': indicators_to_plot,
    'Sum': [sum_female_education, sum_male_education]
})

# Map the 'Indicator' values to 'Female' and 'Male'
df_sums['Indicator'] = df_sums['Indicator'].map({
    indicators_to_plot[0]: 'Female',
    indicators_to_plot[1]: 'Male'
})

# Plot bar chart for the sums
df_sums.plot(kind='bar', x='Indicator', y='Sum', legend=False, colormap='viridis')
plt.title("")
plt.xlabel("Gender")
plt.ylabel("Sum of Values")
plt.show()

```

# Start-up procedures to register a business

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


file_path = 'average_by_country.csv'



columns_to_read_2 = ['Start-up procedures to register a business, female (number)', 'Start-up procedures to register a business, male (number)']

# Read the CSV file with the specified columns
Employers = pd.read_csv(file_path, usecols=columns_to_read_2)

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=Employers, palette="Set2")

# Add labels and title
plt.title('Distribution of Start-up procedures to register a business')
plt.ylabel('%')

# Show the plot
plt.show()




```

# Time required to start a business

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


file_path = 'average_by_country.csv'


columns_to_read = ['Time required to start a business, female (days)', 'Time required to start a business, male (days)']

# Read the CSV file with the specified columns
startup_procedures_df = pd.read_csv(file_path, usecols=columns_to_read)

# Set the style of seaborn
sns.set(style="darkgrid")

# Create a swarm plot for each gender
plt.figure(figsize=(12, 6))
sns.swarmplot(data=startup_procedures_df, palette="Set2")

# Add labels and title
plt.title('Time required to start a business')
plt.ylabel('Time')

# Show the plot
plt.show()

```
