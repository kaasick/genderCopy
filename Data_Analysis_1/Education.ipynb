# Expected years of schooling

```python


# Select a subset of indicators for the plot
indicators_to_plot = [
    
    'Expected years of schooling, female',
    'Expected years of schooling, male'

]

# Plot horizontal bar chart for selected indicators
df_plot = df.set_index('country_name')[indicators_to_plot]
df_plot.plot(kind='barh', figsize=(12, 10), colormap='viridis')
plt.title("Gender Equality Indicators Across Countries")
plt.xlabel("Average Indicator Value")
plt.ylabel("Country")
plt.legend(title="Indicators", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()



# Select indicators for the plot
indicators_to_plot = [
    'Expected years of schooling, female',
    'Expected years of schooling, male'
]

# Sum values across all countries for each indicator
sum_female_education = df[indicators_to_plot[0]].sum()
sum_male_education = df[indicators_to_plot[1]].sum()

# Create a DataFrame for the sums
df_sums = pd.DataFrame({
    'Indicator': indicators_to_plot,
    'Sum': [sum_female_education, sum_male_education]
})

# Plot bar chart for the sums
ax = df_sums.plot(kind='bar', x='Indicator', y='Sum', legend=False, colormap='viridis')
plt.title("Sum of Expected years of schooling")
plt.xlabel("Indicator")
plt.ylabel("Sum of Values")

# Change the x-axis labels
ax.set_xticklabels(['Female', 'Male'], rotation=0)

plt.show()


```

# Female share of graduates

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv("average_by_country.csv")

# Select indicators for the plot
stem_programs = "Female share of graduates from Science, Technology, Engineering and Mathematics (STEM) programmes, tertiary (%)"
arts_programs = "Female share of graduates in Arts and Humanities programmes, tertiary (%)"
business_programs = "Female share of graduates in Business, Administration and Law programmes, tertiary (%)"
engineering_programs = "Female share of graduates in Engineering, Manufacturing and Construction programmes, tertiary (%)"
health_programs = "Female share of graduates in Health and Welfare programmes, tertiary (%)"
ict_programs = "Female share of graduates in Information and Communication Technologies programmes, tertiary (%)"
services_programs = "Female share of graduates in Services programmes, tertiary (%)"
social_sciences_programs = "Female share of graduates in Social Sciences, Journalism and Information programmes, tertiary (%)"
other_programs = "Female share of graduates in other fields than Science, Technology, Engineering and Mathematics programmes, tertiary (%)"

# Select a subset of indicators for the plot
indicators_to_plot = [
    stem_programs,
    arts_programs,
    business_programs,
    engineering_programs,
    health_programs,
    ict_programs,
    services_programs,
    social_sciences_programs,
    other_programs
]

# Plot bar chart for selected indicators
df_plot = df.set_index('country_name')[indicators_to_plot]
df_plot.plot(kind='bar', figsize=(14, 8), colormap='viridis')
plt.title("Female Share of Graduates Across Tertiary Programs")
plt.xlabel("Country")
plt.ylabel("Female Share of Graduates (%)")
plt.legend(title="Tertiary Programs", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.show()
```

# Percentage of students in upper secondary vocational education who are female (%)

```python


# Select indicator for the plot
indicator_to_plot = "Percentage of students in upper secondary vocational education who are female (%)"

# Calculate average across all countries
average_value = df[indicator_to_plot].mean()

# Plot line chart for the selected indicator
df_plot = df.set_index('country_name')[indicator_to_plot]
df_plot.plot(kind='line', figsize=(14, 8), marker='o', colormap='viridis', linestyle='-', label='Individual Countries')
plt.axhline(y=average_value, color='r', linestyle='--', label='Average Across Countries')

plt.title("Percentage of Female Students in Upper Secondary Vocational Education Across Countries")
plt.xlabel("Country")
plt.ylabel("Percentage of Female Students")
plt.legend()
plt.show()


```
